Q2)  Build a simple webpage that helps users navigate different web development-related websites. Note: On clicking the hyperlink the web pages should open in a new tab. Below is a reference image.

Ans)


```python
from IPython.display import display, HTML

html_code = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Assignment-4:- Simple Webpage </title>
</head>
<body>
    <h1>Navigate Me:</h1>
    <div class="pw">
        <p>
            Take me to  <a target="_blank" href="https://www.pwskills.com">PW Skills</a> to buy a course.

        </p>
    </div>
    <div class="mdn">
        <p>
            Take me to  <a target="_blank" href="https://developer.mozilla.org/en-US/">MDN Docs</a> to Know more about Web Development.

        </p>

    </div>
    <div class="lab">
        <p>
            Take me to  <a target="_blank" href="https://pwskills.com/">PW Skills Lab</a> to practice live coding.

        </p>

    </div>
</body>
</html>
"""
display(HTML(html_code))
```
